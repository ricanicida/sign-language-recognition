```python
def download_yt_video(link, file_name, output_path):
    
    youtubeObject = YouTube(link)
    youtubeObject = youtubeObject.streams.get_highest_resolution()
    try:
        youtubeObject.download(output_path=output_path, filename=file_name)
        print("Download is completed successfully")
        return 1
    except:
        print("An error has occurred")
        return 0
```

```python
def create_dataset(df, sign_list, temp_foler_path, dataset_folder_path):
    temp_foler_path = pathlib.Path(temp_foler_path)
    dataset_folder_path = pathlib.Path(dataset_folder_path)


    for sign in sign_list:
        for link in df.loc[df.clean_text == sign]['url'].to_numpy():
            id_list = list(df.loc[(df['url'] == link) & (df['clean_text'].isin(sign_list))].index)
            
            if len(id_list) == 1:
                df_aux = df.iloc[id_list[0]]
                file_name = '-'.join([sign, str(df_aux.label), str(df_aux.signer_id), str(id_list[0])]) + '.mp4'

                try:
                    success = download_yt_video(link, file_name, temp_foler_path)
                except:
                    success = False
                
                if success:
                    # print('The video was successfully downloaded')
                    yt_video = VideoFileClip(str(temp_foler_path / file_name))

                    start_time = df_aux.start_time
                    end_time = df_aux.end_time

                    if not (dataset_folder_path / sign).exists():
                        os.mkdir(dataset_folder_path / sign)
                    
                    if (dataset_folder_path / sign / file_name).exists(): 
                        print('Video already exists')
                    else:
                        try:
                            clip = yt_video.without_audio().subclip(start_time, end_time)
                            clip.write_videofile(str(dataset_folder_path / sign / file_name), audio=False, verbose=False, logger=None)
                            print('Video was successfully saved')
                        except:
                            print('An error occurred while saving the video')
                    yt_video.close()

                    os.remove(temp_foler_path / file_name)
                
                else:
                    print('An error occurred while downloading video from youtube')
            
            elif len(id_list) > 1:
                df_aux = df.iloc[id_list]
                file_name = 'multiple-signs-video.mp4'

                try:
                    success = download_yt_video(link, file_name, temp_foler_path)
                except:
                    success = False

                if success:
                    # print('The video was successfully downloaded')
                    yt_video = VideoFileClip(str(temp_foler_path / file_name))

                    for id in df_aux.index:
                        data = df.iloc[id]
                        clean_text = str(data.clean_text)
                        sign_file_name = '-'.join([clean_text, str(data.label), str(data.signer_id), str(id)]) + '.mp4'

                        start_time = data.start_time
                        end_time = data.end_time

                        if not (dataset_folder_path / clean_text).exists():
                            os.mkdir(dataset_folder_path / clean_text)
                        
                        if (dataset_folder_path / clean_text / sign_file_name).exists(): 
                            print('Video already exists')
                        else:
                            try:
                                clip = yt_video.without_audio().subclip(start_time, end_time)
                                clip.write_videofile(str(dataset_folder_path / clean_text / sign_file_name), audio=False, verbose=False, logger=None)
                                print('Video was successfully saved')
                            except:
                                print('An error occurred while saving the video')
                                
                    yt_video.close()

                    os.remove(temp_foler_path / file_name)
                
                else:
                    print('An error occurred while downloading video from youtube')

                
    return
```

```python
signs = ['eat', 'nice', 'want', 'teacher', 'orange', 'white', 'what', 'like', 'friend', 'fish']
```
