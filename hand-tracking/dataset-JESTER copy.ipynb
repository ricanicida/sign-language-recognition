```python
import cv2
from pathlib import Path
import pandas as pd
import numpy as np
import shutil
```

```python
df_labels = pd.read_csv("D:/Documentos/Polito/Thesis/Datasets/jester/20bn-jester-download-package-labels/labels/labels.csv", header=None)
```

```python
df_labels.head()
```

```python
gestures_list = [
    'Thumb Up',
       'Pushing Hand Away',
       'Sliding Two Fingers Down',
       'Zooming In With Two Fingers',
       'No gesture',
       'Swiping Down',
       'Shaking Hand',
       'Swiping Left', 
       'Rolling Hand Forward',
       'Swiping Right',
       'Turning Hand Clockwise',
]
```

## Train

```python
df_train = pd.read_csv("D:/Documentos/Polito/Thesis/Datasets/jester/20bn-jester-download-package-labels/labels/train.csv", sep=";", header=None)
```

```python
df_train.rename(columns={0:'folder', 1:'label'}, inplace=True)
```

```python
df_train.head()
```

```python
df_train_frquency = pd.DataFrame(df_train.label.value_counts())
df_train_frquency.rename(columns={'label':'count'}, inplace=True)
```

```python
df_train_frquency
```

```python
df_train_frquency.index
```

```python
df_train_10 = df_train.loc[(df_train.label.isin(gestures_list))]
```

```python
df_train_10.head()
```

```python
folders_path = Path('D:/Documentos/Polito/Thesis/Datasets/jester/jester_videos/')
videos_path = Path('D:/Documentos/Polito/Thesis/Datasets/jester/jester_10_gestures/train/')

videos_path.mkdir(parents=True, exist_ok=True)

for gesture in gestures_list:
    (videos_path / gesture).mkdir(parents=True, exist_ok=True)
    
for index, row in df_train_10.iterrows():
    
    video_id = row['folder']
    label = row['label']
    filename = str(video_id) + '.mp4'

    source = folders_path / filename
    destination = videos_path / label / filename

    shutil.copy(source, destination)

    print(label)
    print(filename)
    print('\n')
    
    
```

## Validation

```python
df_val = pd.read_csv("D:/Documentos/Polito/Thesis/Datasets/jester/20bn-jester-download-package-labels/labels/validation.csv", sep=";", header=None)
df_val.rename(columns={0:'folder', 1:'label'}, inplace=True)
df_val.head()
```

```python
df_val_frquency = pd.DataFrame(df_val.label.value_counts())
df_val_frquency.rename(columns={'label':'count'}, inplace=True)
```

```python
df_val_frquency
```

```python
df_val_10 = df_val.loc[(df_val.label.isin(gestures_list))]
```

```python
folders_path = Path('D:/Documentos/Polito/Thesis/Datasets/jester/jester_videos/')
videos_path = Path('D:/Documentos/Polito/Thesis/Datasets/jester/jester_10_gestures/validation/')

videos_path.mkdir(parents=True, exist_ok=True)

size = df_val_10.shape[0]
counter = 0

for gesture in gestures_list:
    (videos_path / gesture).mkdir(parents=True, exist_ok=True)
    
for index, row in df_val_10.iterrows():
    counter += 1

    if counter % 200 == 0:
        print(f"{round(100*counter/size, 2)}%")
    
    video_id = row['folder']
    label = row['label']
    filename = str(video_id) + '.mp4'

    source = folders_path / filename
    destination = videos_path / label / filename

    shutil.copy(source, destination)
```
